```python
from matplotlib import pyplot as plt # import matplotlib library
import numpy as np

import matplotlib.patches as mpatches
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
```

```python
points = [(-15, -3), (-9, -1), (-2, 0), (2, 6)] # list of points
np_points = np.array(points) # convert list of points to np.array
```

```python
%matplotlib inline

fig = plt.figure(num=1) # createing the figure
fig.set_figheight(10)
fig.set_figwidth(10)
c = fig.add_subplot(1, 1, 1) # adding the graph 
c.grid() # createing the grid
c.set_aspect('equal') # preserve all the angles
c.set_xlabel("ось ОX") 
c.set_ylabel("ось ОY")

polygon = mpatches.Polygon(xy=points, closed=True, fill=False, linewidth="3.0", color=(.5, .5, .5)) # defining the polygon
c.add_patch(polygon) # adding the polygon


x = np_points[:, 0] # 
y = np_points[:, 1] # taking the scond column and all the rows
# To chooose (x,y) of the i-th point we write: x,y = np_fig[i, :]
c.plot(x, y, lw=.0, marker='o', color="gold")

fig.savefig('ex1.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300, )
```

```python
points = [[-15, -10], [-4, -1], [2, 1], [6, 10]]
points = [points]
points.append([[-13, -4], [-3, 6], [8, 4], [-2, 10]])

points = np.array(points)

n_pol = points.shape[0]
fig, cs = plt.subplots(n_pol, 1)
fig.set_figheight(10)
fig.set_figwidth(10)
    
for idx, pol_p in enumerate(points):
    cs[idx].set_aspect('equal')
    cs[idx].grid()
    cs[idx].set_xlabel("ось OX")
    cs[idx].set_ylabel("ось OY")
    cs[idx].xaxis.set_minor_locator(MultipleLocator(0.1))

    polygon = mpatches.Polygon(xy=pol_p, closed=True, fill=False, linewidth="3.0", color=(.5, .5, .5))
    cs[idx].add_patch(polygon)
    
    x = pol_p[:, 0]
    y = pol_p[:, 1]

    cs[idx].plot(x, y, lw=.0, marker='o', color="gold")

    fig.savefig('ex1_1.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300)
```
