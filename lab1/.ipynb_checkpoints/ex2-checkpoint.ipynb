```python
from matplotlib import pyplot as plt
import numpy as np

import matplotlib.patches as mpatches

%matplotlib inline
```

```python
def plot_regular_polygon(xy, n_edges, edge_length):
    
    fig = plt.figure(num=1, figsize=(10,10))
    c = fig.add_subplot(1, 1, 1)
    c.set_xlabel(r"Ось O$x$")
    c.set_ylabel("Ось O$у$")
    c.set_title("Правильный многоугольник")
    c.set_xlim(left=-10,right=10)
    c.set_ylim(bottom=-10, top=10)
    c.grid()
    
    # https://www.yaklass.ru/p/geometria/9-klass/dlina-okruzhnosti-i-ploshchad-kruga-9241/pravilnye-mnogougolniki-9246/re-983bb30f-8304-4d02-a739-40bb351cb45d    
    arg = np.radians(180/n_edges)
    out_radius = edge_length/(np.sin(arg) * 2)
    in_radius = out_radius * np.cos(arg)

    rpolygon = mpatches.RegularPolygon(xy, n_edges, out_radius, fill=False, linewidth=3)
    out_circle = mpatches.Circle(xy, out_radius, fill=False, color="red",linewidth=1)
    in_circle = mpatches.Circle(xy, in_radius, fill=False, color="blue",linewidth=1)

    path = rpolygon.get_path()
    vert = path.vertices
    trans = rpolygon.get_patch_transform()
    points = trans.transform(vert)

    c.add_patch(rpolygon)
    c.add_patch(out_circle)
    c.add_patch(in_circle)
    c.plot(points[:, 0], points[:, 1],'o', c="r", markersize=10)
    c.plot([xy[0],xy[0]+in_radius])
    c.plot([xy[0],xy[0]+out_radius],[xy[1],xy[1]])
    c.plot(xy[0], lw=.0 ,color="r", marker='o')
    c.text(x=0.1+xy[0], y=0.1+xy[1], s="O", fontstyle="italic", fontsize=20)

    for i in range(len(points[:-1])):
        c.text(x=1.15* points[i, 0], y=1.15 * points[i, 1], s=f"$P_{i}$",verticalalignment='center', horizontalalignment='center', fontstyle="italic", fontsize=20)
        
    fig.savefig('ex2.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300, )
```

```python
plot_regular_polygon(np.array([0, 0]), 8, 4)
```
