```python
from matplotlib import pyplot as plt 
import numpy as np
```

```python
# Построим график. 
a=2
fig = plt.figure(figsize=(10,10))
c = fig.add_subplot(1,1,1)
c.grid()

theta = np.linspace(10e-6, 4*np.pi - 10e-6, 100)
x = a * np.cos(theta)**3
y = a * np.sin(theta)**3

# При изменении параметра а, необходимо вручную переписать производные. 
diffx1 = -6*np.sin(theta)*np.cos(theta)**2
diffx2 = 12*np.sin(theta)**2 * np.cos(theta) - 6*np.cos(theta)**3
diffy1 = 6*np.sin(theta)**2 * np.cos(theta)
diffy2 = -6*np.sin(theta)**3 + 12*np.sin(theta)*np.cos(theta)**2

denominator = diffx1 * diffy2 - diffx2 * diffy1
squared_sum = np.hypot(diffx1, diffy1)**2

ev_x = x - ( diffy1 * squared_sum ) / denominator
ev_y = y + ( diffx1 * squared_sum ) / denominator

c.plot(x, y)
c.plot(ev_x, ev_y)

fig.savefig('ex4.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300, )
```
