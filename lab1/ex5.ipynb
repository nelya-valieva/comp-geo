```python
from matplotlib import pyplot as plt 
import numpy as np
```

```python
fig = plt.figure(figsize=(15,30))

args = np.array([[4, 6, 3, 1, 1, 1, 1],
                 [6, 3, 2, 1, 1, 1, 1],
                 [2, 10, 1.5, 1, 1, 1, 1],
                 [2, 10, -1.5, 1, 1, 1, 1],
                 [2, 44, -0.2, 1, 1, 1, 1],
                 [8, 40, -0.2, 1, 1, 1, 1],
                 [88, 64, 3, 1, 1, 1, 1],
                 [88, 64, -20, 1, 1, 1, 1]])

for i, _ in enumerate(args):
    m1, m2, n1, n2, n3, a, b = args[i]
    fi = np.linspace(0, 4*np.pi, 1000)
    r = ( np.abs(np.cos(m1*fi/4/a))**n2 + np.abs(np.sin(m2*fi/4)/b)**n3 )**( -1/n1 )
    
    x = r * np.cos(fi)
    y = r * np.sin(fi)
    
    c = fig.add_subplot(4,2,i+1)
    c.grid()
    c.plot(x, y, color='green')
    
    # Rings & Bells
    c.set_ylabel(" $y$", fontsize=20)
    c.set_xlabel(" $x$", fontsize=20)
    c.set_title(f"$m_1={m1}$, $m_2={m2}$, $n_1={n1}$,\n $n_2={n2}$, $n_3={n3}$, $a={a}$, $b={b}$", fontsize=20)
    c.tick_params(axis='both', which='major', labelsize=15)

#       
fig.tight_layout()

fig.savefig('ex5.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300, )
```
