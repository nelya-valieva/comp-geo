```python
from matplotlib import pyplot as plt
from matplotlib import patches as mpatches
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import numpy as np
```

```python
# Radius of the circle
r = 3
# Desired number of segments
s = 1000

theta = np.linspace(0, 4*np.pi, s)
x = r * (theta - np.sin(theta))
y = r * (1 - np.cos(theta)) 
dx = r * (1 - np.cos(theta))
dy = r * (np.sin(theta))


# Fig config
fig = plt.figure(figsize=(20,20))
c = fig.add_subplot(1,1,1)

c.set_ylabel(f" $y$", fontsize=15)
c.set_xlabel(f" $x$", fontsize=15)

plt.xticks(np.arange(min(x), max(x)+1, 3.))
plt.yticks(np.arange(min(y), max(y)+1, 3.))

c.xaxis.set_minor_locator(MultipleLocator(1))

c.grid() 
c.set_aspect('equal')

# Define circle
circle = mpatches.Circle([10, r], r, color="blue", fill=False, lw=3)
# Plot the circle
c.add_patch(circle)
# Plot the center of the circle
c.plot(10, r, 'ro', markersize=8)

# Plot the cycloid
c.plot(x, y, color='green', label="", lw=3)

#  30   
ldx = dx[::30]
ldy = dy[::30]
c.quiver(x[::30], y[::30], ldx, ldy, width=.005, minlength=.05, scale=300, label=" ")

c.legend(ncol=1, framealpha=.7, fontsize=10)

fig.savefig('ex3.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300, )
```
