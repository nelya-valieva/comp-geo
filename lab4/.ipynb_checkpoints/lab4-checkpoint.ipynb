```python
import numpy as np
import matplotlib.pyplot as plt

def func(t):
    y = 1 / (1 + np.power(t, 2))
    return np.array([t, y])
```

```python
def fabric_of_quadratic_polynomial(t, P):
    """Производим квадратные трехчлены"""
    assert P.shape == (3, 2), "На вход три точки (x, y) в виде матрицы 3х2"

    T = np.array([np.ones_like(t), t, t**2]).T
    A = np.linalg.solve(T, P)

    def P(t):
        """Квадратный трехчлен"""
        T = np.array([np.ones_like(t), t, t**2]).T
        return np.matmul(T, A)
        # или
        #return np.array([A[0] + A[1] * t + A[2] * t**2 for t in t])
    return P


def fabric_of_cubic_polynomial(t, P):
    """Производим кубические трехчлены"""
    assert P.shape == (4, 2), "На вход четыре точки (x, y) в виде матрицы 4х2"

    T = np.array([np.ones_like(t), t, t**2, t**3]).T
    A = np.linalg.solve(T, P)

    return lambda t: np.matmul(np.array([np.ones_like(t), t, t**2, t**3]).T, A)

def fabric_of_8_polynomial(t, P):
    """ Производим полиномы 8 порядка """
    assert P.shape == (9, 2), " На вход девять точек (x, y) в виде матрицы 9х2 "

    T = np.array([np.ones_like(t), t, t**2, t**3, t**4, t**5, t**6, t**7, t**8]).T
#     print(T.shape)
    A = np.linalg.solve(T, P)

    return lambda t: np.matmul(np.array([ np.ones_like(t), t, t**2, t**3, t**4, t**5, t**6, t**7, t**8 ]).T, A )


def fabric_of_12_polynomial(t, P):
    """ Производим полиномы 12 порядка """
    assert P.shape == (13, 2), " На вход 13 точек (x, y) в виде матрицы 13х2 "

    T = np.array([np.ones_like(t), t, t**2, t**3, t**4, t**5, t**6, t**7, t**8, t**9, t**10, t**11, t**12]).T
#     print("Shape of T", T.shape)
#     print("Shape of P", P.shape)
    A = np.linalg.solve(T, P)

    return lambda t: np.matmul(np.array([ np.ones_like(t), t, t**2, t**3, t**4, t**5, t**6, t**7, t**8, t**9, 
                                        t**10, t**11, t**12]).T, A )
```

```python
ts = np.linspace(-1.9 * np.pi, 1.9 * np.pi, 100)

t = np.linspace(-6, 6, 3)
print(t.shape)
y = 1 / (1 + np.power(t, 2))
P = np.array([t, y]).T
print(P)

pol2d = fabric_of_quadratic_polynomial(t, P)

t = np.linspace(-1.9 * np.pi, 1.9 * np.pi, 4)
y = 1 / (1 + np.power(t, 2))
P = np.array([t, y]).T

pol3d = fabric_of_cubic_polynomial(t, P)

X1, Y1 = pol2d(ts).T
X2, Y2 = pol3d(ts).T
```

```python
fig01 = plt.figure(num=1)
ax01 = fig01.add_subplot(1,1,1)

# ax01.set_aspect('equal')

ax01.plot(X1, Y1, color='red', label='Квадратный полином')
ax01.plot(X2, Y2, color='green', label='Кубический полином')
ax01.plot(*func(ts), color='k', label="Исходная функция")

ax01.legend(loc='best')
fig01.savefig('ex1.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300)
```

### Нарисуем 3 сегмента сплайна Эрмита по заданным точкам и известным касательным векторам

```python
def F(t):
    """Вектор опорных функций"""
    A = np.array([
        [1, 0, 0, 0],
        [0, 1, 0, 0],
        [-3, -2, 3, -1],
        [2, 1, -2, 1]
    ], dtype=np.float64)
    return np.matmul(np.array([np.ones_like(t), t, t**2, t**3]).T, A)


def cspline(t, p0, p1, v0, v1):
    M = np.array([p0, v0, p1, v1])
    return np.matmul(F(t), M)
```

```python
p0 = np.array([ 0.5, 0.5 ])
p1 = np.array([ 1.5, 0.5 ])
p2 = np.array([ 2.0, 1.0 ])
p3 = np.array([ 1.0, 1.0 ])

v0 = np.array([ 0.25, 0.5 ])
v1 = np.array([ 0.25, -0.25 ])
v2 = np.array([ 0.25, -0.25 ])
v3 = np.array([ 0.25, -0.25 ])
```

```python
t = np.linspace(0, 1, 1000)
X1, Y1 = cspline(t, p0, p1, v0, v1).T
X2, Y2 = cspline(t, p1, p2, v1, v2).T
X3, Y3 = cspline(t, p2, p3, v2, v3).T
```

```python
fig = plt.figure(num=0, figsize=(8,8))
ax = fig.add_subplot(1, 1, 1)

ax.plot(X1, Y1)
ax.plot(X2, Y2)
ax.plot(X3, Y3)
ax.plot([p0[0], p1[0], p2[0], p3[0]], [p0[1], p1[1], p2[1], p3[1]], marker='o', lw=0)

ax.arrow(p0[0], p0[1], p0[0]*v0[0]*0.5, p0[1]*v0[1]*0.5, width=0.0001, head_width=0.02, color="black")
ax.arrow(p1[0], p1[1], p1[0]*v1[0]*0.5, p1[1]*v1[1]*0.5, width=0.0001, head_width=0.02, color="black")
ax.arrow(p2[0], p2[1], p2[0]*v2[0]*0.5, p2[1]*v2[1]*0.5, width=0.0001, head_width=0.02, color="black")
ax.arrow(p3[0], p3[1], p3[0]*v3[0]*0.5, p3[1]*v3[1]*0.5, width=0.0001, head_width=0.02, color="black")
ax.grid()
fig01.savefig('ex3.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300)
```

### Нарисуем сегменты сплайна Эрмита, используя равномерный опорный вектор 

```python
def H00(t):
    """Базисный полином Эрмита $h_{00}(t)$"""
    return (2*t + 1)*(t - 1)**2

H00.name = r'$h_{00}(t)$'

def H10(t):
    """Базисный полином Эрмита $h_{10}(t)$"""
    return t*(t - 1)**2

H10.name = r'$h_{10}(t)$'

def H01(t):
    """Базисный полином Эрмита $h_{01}(t)$"""
    return (3 - 2*t)*(t**2)

H01.name = r'$h_{01}(t)$'

def H11(t):
    """Базисный полином Эрмита $h_{11}(t)$"""
    return (t - 1)*(t**2)

H11.name = r'$h_{11}(t)$'


def spline(T, points, vectors):
    res = []
    tau = np.linspace(0, 1, 100) # as far as I understood, tau is just a set of elements from 0 to 1
    for i in range( len(points) - 1 ):
        res.append( H00(tau)*points[i][:, np.newaxis] + H10(tau)*(T[i+1] - T[i])*vectors[i][:, np.newaxis] + 
                   H01(tau)*points[i+1][:, np.newaxis] + H11(tau)*(T[i+1] - T[i])*vectors[i+1][:, np.newaxis] )
    return res
```

```python
T = [0, 1, 2, 3]
τ = np.linspace(0, 1, 100)
points = [p0, p1, p2, p3]
vectors = [v0, v1, v2, v3]

splines = spline(T, points, vectors)

fig = plt.figure(num=0, figsize=(10,10))
ax = fig.add_subplot(1, 1, 1)

X1, Y1 = splines[0]
X2, Y2 = splines[1]
X3, Y3 = splines[2]

ax.grid()
ax.plot(X1, Y1)
ax.plot(X2, Y2)
ax.plot(X3, Y3)
fig01.savefig('ex3_1.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300)
```

### Построим те же сегменты, используя уже аппроксимированные касательные векторы, найденные с помощью тройных конечных разностей.

```python
# Найдём примерные вектора. 

def approximate_vectors(points, T):
    """
    points: np.array of points
    T: python list of times
    """
    approx_v = []
    v0 = (points[1] - points[0]) / (T[1] - T[0])
    v3 = (points[3] - points[2]) / (T[3] - T[2])
    approx_v.append(v0)

    for i in range(1, len(points) - 1):
        v = 1/2 * ( (points[i+1] - points[i])/(T[i+1] - T[i]) + (points[i] - points[i-1])/(T[i] - T[i-1]) )
        approx_v.append(v)

    approx_v.append(v3)
    
    return approx_v

approx_v = approximate_vectors(points, T)

t = np.linspace(0, 1, 1000)
X1, Y1 = cspline(t, points[0], points[1], approx_v[0], approx_v[1]).T
X2, Y2 = cspline(t, points[1], points[2], approx_v[1], approx_v[2]).T
X3, Y3 = cspline(t, points[2], points[3], approx_v[2], approx_v[3]).T
```

```python
fig = plt.figure(num=0, figsize=(10,10))
ax = fig.add_subplot(1, 1, 1)

ax.plot(X1, Y1)
ax.plot(X2, Y2)
ax.plot(X3, Y3)
ax.grid()
ax.plot([ X1[0], X2[0], X3[0], X3[-1] ], [ Y1[0], Y2[0], Y3[0], Y3[-1] ], lw=0, marker='o')
fig01.savefig('ex3_2.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300)
```

```python
# Реализация кардинального сплайна с возможностью изменения параметра 𝑐.
# Повторим рисунки с 11 по 16. Точки имеют целочисленные координаты, поэтому их можно определить из рисунков.

# На что влияет параметр 𝑐? Параметр  с влияет на степень "натяжения" сплайна. Чем больше число с, тем большее
# натяжение мы имеем 

def canonic_approx_vectors(points, T, c):
    """
    points: np.array of points
    T: python list of times
    """
    assert 0 <= c <= 1, "c has to be from 0 to 1, including the borders."
    
    approx_v = []

    for i in range(1, len(points) - 1):
        v = (1 - c) * (points[i + 1]  - points[i - 1]) / (T[i + 1] - T[i - 1])
        approx_v.append(v)
    
    return approx_v

# Рисование
T = [0, 1, 2, 3, 4, 5, 6, 7, 8]
points = np.array([ [0, 1], [2, 1.5], [2, 3], [4, 4], [6, 3], [3, 2], [2, 5], [1, 3], [0, 1] ])
c = [.2, .4, .5, .7, .8]
columns = 3
rows = round( len(c) / columns )

for i in range(1, len(c) + 1):
    fig = plt.figure(num=1, figsize=(16, 8))
    ax = fig.add_subplot(rows, columns, i)
    ax.grid()
    ax.set_aspect('equal')
    ax.set_xlabel("Ось $Ox$")
    ax.set_ylabel("Ось $Oy$")
    ax.set_title(f"Кардинальный сплайн, $c$ = {c[i-1]}")

    approx_vectors = canonic_approx_vectors(points, T, c[i-1])

    splines = spline(T[1:-1], points[1:-1], approx_vectors)

    for sp in splines:
        X, Y = sp
        ax.plot(X, Y)
        ax.plot(X[0], Y[0], marker='o')
fig.savefig('ex4_1.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300)
```

```python
# Используем неравномерный опорный вектор t значений параметра 𝑡. Что изменится в рисунках в этом случае?
# Например, можно использовать хордовую интерполяцию

def points_dist(p1, p0):
    return np.sqrt( (p1[0] - p0[0])**2 + (p1[1] - p0[1])**2 )

def t_approx_chord_interpolation(points):
    T = [0]
    for i in range(1, len(points)):
        T.append(points_dist( points[i], points[i-1] ))
    return T

def canonic_chord_approx_vectors(points, T, c):
    """
    points: np.array of points
    T: python list of times
    """
    assert 0 <= c <= 1, "c has to be from 0 to 1, including the borders."
    
    approx_v = []

    for i in range(1, len(points) - 1):
        v = (1 - c) * (points[i + 1]  - points[i - 1]) / (T[i])
        approx_v.append(v)
    
    return approx_v


# Рисование
points = np.array([ [0, 1], [2, 1.5], [2, 3], [4, 4], [6, 3], [3, 2], [2, 5], [1, 3], [0, 1] ])
T = t_approx_chord_interpolation(points)
print(T)
c = [.2, .4, .5, .7, .8]
columns = 1
rows = round( len(c) / columns )
print(rows)

fig = plt.figure(num=1, figsize=(32, 32))

for i in range(1, len(c) + 1):
    ax = fig.add_subplot(rows, columns, i)
    ax.grid()
    ax.set_aspect('equal')
    ax.set_xlabel("Ось $Ox$")
    ax.set_title(f"Кардинальный сплайн, c = {c[i-1]}")

    approx_vectors = canonic_chord_approx_vectors(points, T, c[i-1])

    splines = spline(T[1:-1], points[1:-1], approx_vectors)

    for i, sp in enumerate(splines):
        X, Y = sp
        ax.plot(X, Y)
        ax.plot(X[0], Y[0], marker='o')
        ax.plot( [X[0], X[0] + approx_vectors[i][0] * .5], [Y[0], Y[0] + approx_vectors[i][1] * .5] )
        
fig.savefig('ex4_2.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300)
```

## Задание № 5.
Реализуйте построения TCB-сплайна. Изучите влияние параметра 𝛾 на непрерывность сплайна.
Для чего изменяйте его с шагом 1/10 от −1 до 1 и для каждого стройте рисунок. Первый рисунок
должен совпадать с рис. 17, а последний с рис. 18. Смотрите также презентацию по данной теме.
Значения TCB-параметров можно взять одинаковыми для всех сегментов, то есть 𝛾𝑖 = 𝛾 = const,
𝜃𝑖 = 𝜃 = const и 𝛽𝑖 = 𝛽 = const ∀𝑖. Значения параметров 𝜃 и 𝛽 можно положить равными нулю:
𝜃 = 𝛽 = 0. Набор точек следующий: 𝑃0 = (0, 0), 𝑃1 = (2, 2), 𝑃2 = (4, 2), 𝑃3 = (6, 0).

```python
# write \theta and press TAB to get the greek letters.
def TCB_spline(θ, γ, β, points, T):
    
    if type(θ) in {float, int}:
        θ = [θ] * len(points)
        γ = [γ] * len(points)
        β = [β] * len(points)
    
    # u is the start vectors, v is the end vectors. 
    v = []
    u = []
    
    # At the start and finish v1 = u1, and vn = un
    v1 = (points[1] - points[0]) / (T[1] - T[0])
    vn = (points[-1] - points[-2]) / (T[-1] - T[-2])
    
    v.append(v1)
    u.append(v1)
    
    for i in range(1,  len(points) - 1):
        vi = (1 - θ[i]) * (1 + γ[i]) * (1 - β[i]) / 2 * (points[i+1] - points[i]) / (T[i+1] - T[i]) + \
            (1 - θ[i]) * (1 - γ[i]) * (1 + β[i]) / 2 * (points[i] - points[i-1]) / (T[i] - T[i-1])
        
        ui = (1 - θ[i]) * (1 - γ[i]) * (1 - β[i]) / 2 * (points[i+1] - points[i]) / (T[i+1] - T[i]) + \
            (1 - θ[i]) * (1 + γ[i]) * (1 + β[i]) / 2 * (points[i] - points[i-1]) / (T[i] - T[i-1])
        
        v.append(vi)
        u.append(ui)
    
    v.append(vn)
    u.append(vn)
    
    res = []
    tau = np.linspace(0, 1, 100) # as far as I understood, tau is just a set of elements from 0 to 1
    for i in range( len(points) - 1 ):
        res.append( H00(tau)*points[i][:, np.newaxis] + H10(tau)*(T[i+1] - T[i])*u[i][:, np.newaxis] + 
                   H01(tau)*points[i+1][:, np.newaxis] + H11(tau)*(T[i+1] - T[i])*v[i+1][:, np.newaxis] )
    return res, u, v


# Рисуем
points = np.array([ [0,0], [2, 2], [4, 2], [6, 0] ])
T = [0, 1, 2, 3]
fig = plt.figure(num=1, figsize=(16, 16))
rows = 5
columns = 2

for i in range(1, 11):
    splines, u, v = TCB_spline(0.0, -1.0 + i * 0.1 , 0.0, points, T)
    
    ax = fig.add_subplot(rows, columns, i)
    ax.grid()

    for sp in splines:
        X, Y = sp
        ax.plot(X, Y)
        ax.plot(X[0], Y[0], marker='o')

    ax.plot(X[-1], Y[-1], marker='o', label=f"γ = {-1.0 + i * 0.1}")
    ax.legend()

fig.savefig('ex5.pdf', format='pdf', bbox_inches='tight', pad_inches=0, dpi=300)
```
